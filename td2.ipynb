```python
def Exercice1():
    liste = []
    final_arr = []
    with open("tableau.dat", "r") as f:
        data = f.read().splitlines()

    for i in data:
        liste.append(i.split(";"))
    for i in range(len(liste)):
        for j in range(len(liste[i])):
            final_arr.append(int(liste[i][j]))
    print(final_arr)

    #Trier la liste final_arr dans l'ordre croissant
    final_arr.sort()
    return final_arr

print(Exercice1())
```

```python
class Point3D:
    def __init__(self, x:int, y:int, z:int):
        self.x = x
        self.y = y
        self.z = z

    def __str__(self):
        return "({},{},{})".format(self.x, self.y, self.z)

    def module(self):
        return (self.x**2 + self.y**2 + self.z**2)**0.5
    
    def __add__(self, other):
        return Point3D(self.x + other.x, self.y + other.y, self.z + other.z)

    def __sub__(self, other):
        return Point3D(self.x - other.x, self.y - other.y, self.z - other.z)
    
    def __equal__(self, other):
        return self.x == other.x and self.y == other.y and self.z == other.z

    def __mul__(self, other):
        if isinstance(other, Point3D):  # Produit vectoriel
            return Point3D(self.y * other.z - self.z * other.y, self.z * other.x - self.x * other.z, self.x * other.y - self.y * other.x)
        elif isinstance(other, (int, float)):  # Multiplication par un scalaire
            return Point3D(self.x * other, self.y * other, self.z * other)
    
    def __setitem__(self, key, value): #Set item
        if key == 0:
            self.x = value
        elif key == 1:
            self.y = value
        elif key == 2:
            self.z = value

    
    def __getitem__(self, key): #Get item
        if key == 0:
            return self.x
        elif key == 1:
            return self.y
        elif key == 2:
            return self.z
        
    #Produit scalaire
    def __matmul__(self, other):
        return self.x*other.x + self.y*other.y + self.z*other.z
    

class Mass3D(Point3D):
    def __init__(self, x=0, y=0, z=0, m=0):
        super().__init__(x, y, z) #Heritage
        self.m = m  # Masse associée

    # Afficher les informations de la masse
    def afficher_masse(self):
        return f"Coordonnées: {self}, Masse: {self.m}"
    
    # Modifier la masse
    def set_masse(self, m):
        self.m = m

    # Accéder à la masse
    def get_masse(self):
        return self.m


#Test
print("----------- TEST VECTEUR -----------")
p1 = Point3D(1, 2, 3)
p2 = Point3D(4, 5, 6)
print("P1 :",p1)
print("P2 :",p2)
print("Module de P1",p1.module())
print("P1 + P2",p1 + p2)
print("P1 - P2",p1 - p2)
print("isEqual",p1 == p2)
print("P1 x P2",p1 * p2)
print("P1 x 2",p1 * 2)
p1[0] = 2
print("Modification de la coordonnée x de P1",p1[0])
print("Produit scalaire de P1 et P2", p1 @ p2)
print("----------- TEST MASSE -----------")
m1 = Mass3D(1, 2, 3, 4)
print(m1.afficher_masse())
m1.set_masse(10)
print("Modification de la masse m de m1,",m1.get_masse())
print("Masse de m1",m1.get_masse())
print("Coordonnées de m1",m1)
print("Module de m1",m1.module())
```
